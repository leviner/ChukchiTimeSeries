```python
tr17B = .5
tr17 = np.arange(67,72.6,tr17B)
tr19B = .75
tr19 = np.arange(66.5,72.6,tr19B)

fig, (ax1,ax2) = plt.subplots(ncols=2,figsize = (9,9))
for df,survey in zip([df2017Summary,df2019Summary],[201701,201901]):
    meanA = []
    meanP = []
    if survey == 201701:
        tr = tr17
        trB = tr17B
    if survey == 201901:
        tr = tr19
        trB = tr19B
    for i in range(len(tr)):
        meanP.append(df[(df.Latitude > (tr[i]-trB)) & (df.Latitude < (tr[i]+trB))]['Pollock (No. m^-2)'].mean())
        meanA.append(df[(df.Latitude > (tr[i]-trB)) & (df.Latitude < (tr[i]+trB))]['Arctic cod (No. m^-2)'].mean())
    if survey == 201701:
        ax1.plot(meanP, tr17, 'o',color='#800080')
        ax1.plot(meanA, tr17, 'o',color='#008000')
    if survey == 201901:
        ax2.plot(meanP, tr19,'o',color='#800080')
        ax2.plot(meanA, tr19,  'o',color='#008000')
ax1.set_xlabel('Abundance (No. m^-2)')
ax2.set_xlabel('Abundance (No. m^-2)')
ax1.set_ylabel('Latitude')
ax2.set_ylabel('Latitude')
```
