```python
import pandas as pd
from glob import glob
import sys
import numpy as np
sys.path.insert(1, 'C:/Users/Robert/Documents/projects/rltools')
#sys.path.insert(1, 'C:/Users/robert.levine/work/repositories/rltools')
from akMaps import akMaps
mapper = akMaps.akMaps()
import matplotlib.pyplot as plt
import matplotlib.colors as clr
import cartopy.crs as ccrs
df2012Summary = pd.read_csv('../data/acousticData/2012_2013/Arctic_EIS_Acoustic_trawl_survey_alongtrack_summary_2012_v2.csv') 
df2012Summary['Pollock (No. m^-2)']= 0
df2013Summary = pd.read_csv('../data/acousticData/2012_2013/Arctic_EIS_Acoustic_trawl_survey_alongtrack_summary_2013_v2.csv')
df2013Summary['Pollock (No. m^-2)']= 0
df2017Summary = pd.read_csv('../data/catchData/2017_2019/Arctic_EIS_Acoustic_trawl_survey_alongtrack_summary_2017.csv')
df2019Summary = pd.read_csv('../data/catchData/2017_2019/Arctic_EIS_Acoustic_trawl_survey_alongtrack_summary_2019.csv')
```

```python
cmapAcod = clr.LinearSegmentedColormap.from_list('acod',['#808080','#008000','#00b300','#00ff00'], N=256)
figure = plt.figure(figsize=(20,10),dpi=150)
ax=plt.subplot(141,projection=ccrs.Mercator())
mapper.chukchiScatter(df2012Summary.Latitude, df2012Summary.Longitude, np.log10(df2012Summary['Arctic cod (No. m^-2)']*100),3,extent=[-155,-171,65.5,73.5],ax=ax,colormap=cmapAcod)
ax.set_title('2012 Arctic cod')
ax=plt.subplot(142,projection=ccrs.Mercator())
mapper.chukchiScatter(df2013Summary.Latitude, df2013Summary.Longitude, np.log10(df2013Summary['Arctic cod (No. m^-2)']*100),3,extent=[-155,-171,65.5,73.5],ax=ax,colormap=cmapAcod)
ax.set_title('2013 Arctic cod')
ax=plt.subplot(143,projection=ccrs.Mercator())
mapper.chukchiScatter(df2017Summary.Latitude, df2017Summary.Longitude, np.log10(df2017Summary['Arctic cod (No. m^-2)']*100),3,extent=[-155,-171,65.5,73.5],ax=ax,colormap=cmapAcod)
ax.set_title('2017 Arctic cod')
ax=plt.subplot(144,projection=ccrs.Mercator())
mapper.chukchiScatter(df2019Summary.Latitude, df2019Summary.Longitude, np.log10(df2019Summary['Arctic cod (No. m^-2)']*100),3,extent=[-155,-171,65.5,73.5],ax=ax,colormap=cmapAcod)
ax.set_title('2019 Arctic cod')
plt.subplots_adjust(wspace=.3)
```
