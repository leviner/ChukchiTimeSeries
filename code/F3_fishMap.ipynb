```python
files17 = glob('../data/acousticData/2017_2019/EV/2017/echoview/exports/1m/*(cells).csv')
files19 = glob('../data/acousticData/2017_2019/EV/2019/echoview/exports/1m/*(cells).csv')

def readExports(files):
    dfs = []
    for file in files:
        dfs.append(pd.read_csv(file))
    dfA = pd.concat(dfs)
    dfA = dfA[dfA.Region_class!= 'SURFTURB']
    return dfA

df17 = readExports(files17)
df19 = readExports(files19)    
```

```python
files17 = glob('../data/acousticData/2017_2019/EV/2017/echoview/exports/1m/*(intervals).csv')
files19 = glob('../data/acousticData/2017_2019/EV/2019/echoview/exports/1m/*(intervals).csv')

def readExportsI(files):
    dfs = []
    for file in files:
        dfs.append(pd.read_csv(file))
    dfA = pd.concat(dfs)
    #dfA = dfA.rename(columns={'Process_ID':'PROCESS_ID','Interval':'INTERVAL'})
    return dfA

df17i = readExportsI(files17)
df19i = readExportsI(files19)    
```

```python
from geopy.distance import distance
a = pd.read_csv('catchAnalysis/analysisFiles/dfStat.csv')
a = a[a.Year == 2019][['latitude','longitude','T_surf','T_bot']]
a['longitude'] = (360-a.longitude)*-1
ss,sb = [],[]
for indexi, rowi in df19i.iterrows():
    dcur=1000
    for indexc, rowc in a.iterrows():
        try:
            try:
                if distance((rowi.Lat_E,rowi.Lon_E),(rowc.latitude,rowc.longitude)).km < dcur:
                    dcur = distance((rowi.Lat_E,rowi.Lon_E),(rowc.latitude,rowc.longitude)).km
                    sscur = rowc.T_surf
                    sbcur = rowc.T_bot
            except:
                if distance((rowi.Lat_S,rowi.Lon_S),(rowc.latitude,rowc.longitude)).km < dcur:
                    dcur = distance((rowi.Lat_E,rowi.Lon_E),(rowc.latitude,rowc.longitude)).km
                    sscur = rowc.T_surf
                    sbcur = rowc.T_bot
        except:
            dcur = 1000
            sscur = np.nan
            sbcur = np.nan
    ss.append(sscur)
    sb.append(sbcur)
df19i['T_surf'] = ss
df19i['T_bot'] = sb
df19A = pd.merge(df19, df19i,  how='left', left_on=['Process_ID', 'Interval'], right_on =['Process_ID', 'Interval'])
```
